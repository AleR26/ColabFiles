 ## 1. Operaciones básicas
 Objetivo: Crear y manipular arrays.
 Instrucciones:
 1. Crea un array de 10 elementos con valores entre 1 y 10.
 2. Calcula la media, mediana, y desviación estándar de los valores.
 3. Multiplica cada valor por 2.

```python
import numpy as np
import random

arr = np.random.randint(1,10, size = 10)
print("Array de 10 elementos: ", arr)
media = arr.mean()
mediana = np.median(arr)
desv_est = arr.std()
print("Media: ", media)
print("Mediana: ", mediana)
print("Desviación estándar ", desv_est)
```

## 3. Estadística descriptiva
 Objetivo: Aplicar análisis estadístico básico.
 Instrucciones:
 1. Crea un DataFrame con 100 valores aleatorios entre 0 y 50.
 2. Calcula la media, mediana y moda de los valores.
 3. Encuentra el valor mínimo y el valor máximo del DataFrame

```python
# Crear el DataFrame
df_100 = pd.DataFrame(
    np.random.randint(0, 50, size=100)
)
print("Dataframe\n",df_100.head())
media_df = df_100.mean()[0]
median_df = df_100.median()[0]
mode_df = df_100.mode()[0]

print("Promedio: ", media_df)
print("Mediana: ", median_df)
print("Moda: ", mode_df)

maxi = df_100.max()[0]
mini = df_100.min()[0]
print("Valor Máximo: ", maxi)
print("Valor Mínimo: ", mini)

```

 ## 9. Heatmap
 Objetivo: Visualizar la correlación entre variables numéricas.
 Instrucciones:
 1. Usa el dataset iris de Seaborn.
2. Calcula la matriz de correlacin de las variables.
 3. Crea un heatmap para visualizar la matriz de correlación

```python
from sklearn.preprocessing import OneHotEncoder

iris = sns.load_dataset('iris')
species = list(iris['species'].unique())
mapping = {}
for i in range(len(species)):
  mapping[species[i]] = i
mapping
iris['species_map'] = iris['species'].map(mapping)
iris = iris.drop(columns=['species'])

sns.heatmap(iris.corr(), annot=True)
```

##10. Gráfico de Línea con Media Móvil
 Objetivo: Visualizar series temporales.
 Instrucciones:
 1. Genera una serie temporal de 100 días con valores aleatorios.
 2. Calcula la media móvil de los últimos 10 días y gráficala junto a la serie
 original.
 3. Usa diferentes colores para la serie original y la media móvil

```python
df_temporal = pd.DataFrame({'Fecha': pd.date_range(start='7/20/2025', end='10/27/2025', freq='d'), 'Valor': np.random.randint(20, 40, 100)})
df_temporal['movil'] = df_temporal['Valor'].rolling(window=10, center=False).mean()

plt.plot(df_temporal['Fecha'], df_temporal['Valor'], label='Valor')
plt.plot(df_temporal['Fecha'], df_temporal['movil'], label='Media Móvil')
plt.title("Serie de Tiempo")

plt.xlabel("Fecha")
plt.ylabel("Valor")
plt.grid(True, linestyle=':')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

 ## 11. Datos Temporales con Análisis Estadístico
 Objetivo: Trabajar con series temporales y calcular estadísticas.
 Instrucciones:
 1. Genera una serie temporal de 200 días con valores aleatorios.
 2. Calcula la media y la desviación estándar de la serie temporal.
 3. Visualiza la serie junto con las bandas de desviación estándar (es decir, la
 media ± 1 desviación estándar).

```python
df_temporal = pd.DataFrame({'Fecha': pd.date_range(start='4/11/2025', end='10/27/2025', freq='d'), 'Valor': np.random.randint(20, 40, 200)})

mean = df_temporal["Valor"].mean()
std = df_temporal["Valor"].std()

plt.plot(df_temporal['Fecha'], df_temporal['Valor'], label='Valor', alpha=0.8)
plt.axhline(mean + std, linestyle='--', label='+1σ', color='orange')
plt.axhline(mean, linestyle='--', label='promedio', color='green')
plt.axhline(mean - std, linestyle='--', label='-1σ', color='orange')
plt.title("Serie de Tiempo")

plt.xlabel("Fecha")
plt.ylabel("Valor")
plt.grid(True, linestyle=':')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45)
plt.show()
```
