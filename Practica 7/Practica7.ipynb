#Practica #7. An√°lisis de un Dataset con Matplotlib

---



```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df.head()
```

```python
df.info()
```

Grafico de barras

```python
import pandas as pd

df = pd.DataFrame(df)
mean_species = df.groupby('species').agg(
    {
        'petal_length' : 'mean',
        'petal_width' : 'mean'
    }
)

display(mean_species)
mean_species.plot(kind = 'bar')
plt.title('Promedio de longitud de los petalos por especie')
plt.xlabel('Especies')
plt.ylabel('Mean')
```

Histograma

```python
plt.hist(df['sepal_length'], bins=10, color = 'purple')
plt.title('Distribucion de longitud de los sepalos')
plt.xlabel('Longitud')
plt.ylabel('Frecuencia')
plt.show()
```

Gr√°fico de dispersi√≥n (scatter plot)

```python
x = df.loc[:, 'petal_length']
y = df.loc[:,'petal_width']

# Create separate scatter plots for each species
plt.scatter(x[df['species'] == 'setosa'], y[df['species'] == 'setosa'], c='red', label='setosa')
plt.scatter(x[df['species'] == 'versicolor'], y[df['species'] == 'versicolor'], c='blue', label='versicolor')
plt.scatter(x[df['species'] == 'virginica'], y[df['species'] == 'virginica'], c='green', label='virginica')

plt.title('Largo vs Ancho de P√©talos')
plt.xlabel('Largo de P√©talos (cm)')
plt.ylabel('Ancho de P√©talos (cm)')
plt.legend()
plt.show()
```

Gr√°fico de cajas (box plot)

```python
x = df.loc[:, 'sepal_length']
y = df.loc[:,'sepal_width']


plt.boxplot([x[df['species'] == 'setosa'],x[df['species'] == 'versicolor'],x[df['species'] == 'virginica'],y[df['species'] == 'setosa'],y[df['species'] == 'versicolor'],y[df['species'] == 'virginica']],
            tick_labels= ['long. setosa', 'versicolor','virginica','ancho setosa','versicolor','virginica'])
plt.ylabel('')
plt.show()
```

Gr√°fico de l√≠neas

```python
df['petal_length'].plot(kind = 'line')
plt.plot(df['petal_length'],df['species'])
plt.title('Promedio de longitud de los petalos por especie')
plt.xlabel('Especies')
plt.ylabel('Mean')
```
