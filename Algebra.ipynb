```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/ClassFiles/datos_academicos.xls')
df
```

```python
lst= []
lst = df.columns[2:]
lst
```

```python
promedio = 0
mean_lst = []
for i in lst:
  x = df['Peso'] * df[i]
  promedio = x.sum()
  mean_lst.append(promedio)
print(mean_lst)
```

```python
prom = pd.DataFrame(mean_lst, columns=['Promedio'])
promedio = pd.DataFrame(lst, columns=['Alumnos'])
promedio['Promedio'] = prom
promedio
```

```python
x = []
for i in lst:
  x.append(df[i].mean())
x
```

```python
promedio2 = pd.DataFrame(x, columns=['Promedio 2'])
promedio['Promedio2'] = promedio2
promedio
```

```python
import matplotlib.pyplot as plt

promedio = promedio.groupby('Alumnos').agg(
    {
        'Promedio' : 'mean',
        'Promedio2' : 'mean'
    }
)
promedio.plot(kind = 'bar')
plt.title('Comparativo de promedios')
plt.xlabel('Alumnos')
plt.ylabel('Mean')
```
