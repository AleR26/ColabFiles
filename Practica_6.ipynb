 Análisis de la distribución de supervivencia por combinación de sexo y clase del pasajero.

```python
total = df['Age'].count()
#agrupar por clase, supervivencia y sexo para saber el total de cada grupo
sex_class = df.groupby(['Pclass','Sex','Survived']).agg({
    'Name' : 'count'
})
comb = (sex_class/total)*100
comb = comb.reset_index()
print(comb)

#Calculamos la mayor taza de supervivencia
supervivencia_alta = comb.loc[comb['Survived']== 1]
supervivencia_alta = comb.loc[supervivencia_alta['Name'].idxmax()]
print(f"Las personas de sexo {supervivencia_alta['Sex']} de la clase {supervivencia_alta['Pclass']} fueron las que tuvieron mayor taza de supervivencia.")

#Calculamos la menor taza de supervivencia
supervivencia_baja = comb.loc[comb['Survived']== 0]
supervivencia_baja = comb.loc[supervivencia_baja['Name'].idxmax()]
print(f"Las personas de sexo {supervivencia_baja['Sex']} de la clase {supervivencia_baja['Pclass']} fueron las que tuvieron menor taza de supervivencia.")
```

Identificación de familias grandes a bordo.

```python
df['FamilySize'] = df[['Parents/Children Aboard','Siblings/Spouses Aboard']].sum(axis = 1)
familia_grande = df.loc[df['FamilySize'] > 3]
#El numero de pasajeros en familias grandes se muestra en el FamilySize
print(familia_grande)


super_familia_grande = familia_grande.groupby(['Survived']).agg({
    'FamilySize' : 'sum'
})
#Proporcion de supervivencia por familias grandes
comb = (super_familia_grande/total)*100
comb = comb.reset_index()
print(comb)

```
