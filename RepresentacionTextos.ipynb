```python
def frecuency(corpus):
  tfVec = TfidfVectorizer()
  X = tfVec.fit_transform(corpus)

  Tfid_df = pd.DataFrame(
      X.toarray(),
      columns = tfVec.get_feature_names_out()
  )
  return Tfid_df,tfVec
  print(Tfid_df)
```

```python
def preprocess_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'\W+', ' ', texto)
    return texto

def tokenization(df):
    nlp = spacy.load('es_core_news_sm')
    docs = nlp(df)
    tokens = [token.text for token in docs]
    print("Tokenización")
    print(tokens)
    stop_words  = set(nltk.corpus.stopwords.words('spanish'))
    tokens_clean = [word for word in tokens if word not in stop_words]
    print("Sin stopwords")
    print(tokens_clean)
    return tokens_clean

def lematizacion(tokens_list):
  nlp = spacy.load('es_core_news_sm')
  text = ' '.join(tokens_list)
  docs = nlp(text)
  lemmas = [token.lemma_ for token in docs if not token.is_punct and not token.is_space]
  print(lemmas)
  return lemmas
```

```python
preprocessed_corpus = [preprocess_texto(text) for text in corpus_es]
joined_preprocessed_text = " ".join(preprocessed_corpus)

tokens = tokenization(joined_preprocessed_text)
lemas = lematizacion(tokens)
print(preprocessed_corpus)
```
